```python
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('fig/airplane.bmp')
mask = cv2.imread('fig/mask_plane.bmp', cv2.IMREAD_GRAYSCALE) # 흑백으로 가져오기
dst = cv2.imread('fig/field.bmp')

if src is None or mask is None or dst is None:  # 이미지가 없으면 없다고 출력 후 이미지 뛰우기 종료
    print(' image not found')
    sys.exit()

cv2.copyTo(src,mask,dst)  # (무엇을, 어떡해, 어디로) 이미지 옮겨 만들기


cv2.imshow('src',src)
cv2.imshow('mask', mask)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread('fig/cow.png')
dst = cv2.imread('fig/green.png')

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)  # 원하는 이미지를 어떤(흑백) 형식으로 바꿈
retval, mask = cv2.threshold(src_gray,245,255, cv2.THRESH_BINARY_INV) # 이미지에서 원하는 밝기 (num, num)의 값 사이를 어떤 형식으로 자름 

h,w = src.shape[:2]  # 첫 2개가 (세로, 가로)
dst_re = cv2.resize(dst, (w,h), interpolation= cv2.INTER_AREA)  # resize는 (가로, 세로)로 입력

cv2.copyTo(src, mask, dst_re)

if src is None or dst is None:
    print(' image not found')
    sys.exit()

cv2.imshow('src',mask)
cv2.imshow('dst', dst_re)
cv2.imshow('gray', src_gray)


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('fig/puppy.bmp')

if img is None:
    print('image not found')
    sys.exit()
    
cv2.namedWindow('image', cv2.WINDOW_FULLSCREEN)  # 하나의 창을 'image'라는 이름으로 설정 -> 이 창의 종류 지정
cv2.imshow('image',img)   # '이름' 을 가상의 창 'image'로 지정함으로 위에서 만든 창 불러옴

while True:  # ESC 또는 q를 누르면 종료로 설정
    key = cv2.waitKey()
    if key == 27 or key == ord('q'):
        break

cv2.destroyAllWindows() # 이미지 닫기?
```

```python
src = cv2.imread('fig/cow.png')
dst = cv2.imread('fig/green.png')

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY) # 원본 이미지를 cv2.지정하는 이미지로 변경 - 흑백, RGB, 자외선 등..
retval, mask = cv2.threshold(src_gray,240,255, cv2.THRESH_BINARY_INV)

h,w = src.shape[:2]
# dst_re = cv2.resize(dst, (w,h), interpolation= cv2.INTER_AREA)

dst_crop = dst[350: 350+h, 500:500 + w]     #원하는 값 만큼만 이미지 자르기

cv2.copyTo(src, mask, dst_crop)

if src is None or dst is None:
    print(' image not found')
    sys.exit()

cv2.imshow('src',mask)
cv2.imshow('dst', dst)
cv2.imshow('gray', src_gray)


cv2.waitKey()
cv2.destroyAllWindows()

```

```python
import glob
img_paths = glob.glob('./fig/*.*')

# display(img_paths)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image',cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)


if img_paths is None:
    print('image not found')

idx = 0
while True:
    img = cv2.imread(img_paths[idx])
    
    if img is None:
        print('image not found')
        break
    
    cv2.imshow('image', img)
    
    if  cv2.waitKey(3000) == 27:
        break
    idx += 1

    if idx >= (len(img_paths)-3):
        idx = 0

cv2.destroyAllWindows()
```
