```python
import sys, cv2
import numpy as np
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('cam  not found')
    sys.exit()
    
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h= int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('fig/output.avi',fourcc, fps,(w,h), isColor=False)


while True:
    ret, frame = cap.read()
    if not ret:
        print('frame not found')
        break
    
    edge = cv2.Canny(frame, 80,120)
    
    out.write(edge)
    
    cv2.imshow('frame', edge)
    
    if cv2.waitKey(20) == 27:
        break


out.release()
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('cam  not found')
    sys.exit()

# cv2.imshow('frame', frame)

f_key = False
e_key = True
i_key = False
b_key = False     # 스위치 설정

while True:
    ret, frame = cap.read()
    # frame = frame.copy()
    key = cv2.waitKey(20)
    if not ret:
        print('frame not found')
        break
    
    # cv2.imshow('frame', frame) # 실행 즉시 영상 뜸
    
    if f_key:
        frame = cv2.flip(frame, 1)

    if b_key:
        frame = cv2.flip(frame, 0)    

    if e_key:
        frame = cv2.Canny(frame, 50, 100)
    
    if i_key:
        frame = 255 - frame
    
    cv2.imshow('frame', frame)  # 변경된 설정 적용 후 출력
    
    if key == 27 or key == ord('q'):
        break
    
    elif key == ord('e'):   # 스위치 변경 트리거
        e_key = not e_key
    
    elif key == ord('f'):
        f_key = not f_key
        
    elif key == ord('i'):
        i_key = not i_key
        
    elif key == ord('b'):
        b_key = not b_key
        
out.release()
cap.release()
cv2.destroyAllWindows()
```

```python

img = np.ones((600,600,3),np.uint8)*255
img1 = img.copy()

cv2.namedWindow('img')
cv2.setMouseCallback('img', click)

while True:
    cv2.imshow('img',img1)
    key = cv2.waitKey()
    if key == 27 or key == ord('q'):
        break    
    
    if key == ord('s'):
        print('image saved')
        cv2.imwrite('fig/save.png', img)
    
    elif key == ord('c'):
        img = img1.copy()
        cv2.imshow('img',img)

cv2.destroyAllWindows()
```
